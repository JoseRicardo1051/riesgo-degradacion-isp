```python
# Excluimos cliente_id porque es un identificador
corr_df = df.drop(columns=["cliente_id"], errors="ignore")

corr_matrix = corr_df.corr(numeric_only=True)

plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Matriz de correlación")
plt.show()
```

```python
corr_target = corr_matrix[target].drop(target).sort_values(key=abs, ascending=False)
print("Correlación con la variable objetivo:")
display(corr_target)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=corr_target.values, y=corr_target.index)
plt.title("Correlación de variables con la variable objetivo")
plt.xlabel("Correlación")
plt.ylabel("Variable")
plt.show()
```

```python
features_num = [col for col in num_cols if col not in ["cliente_id", target]]

for col in features_num:
    plt.figure(figsize=(8,4))
    sns.boxplot(data=df, x=target, y=col)
    plt.title(f"{col} vs {target}")
    plt.show()
```

```python
for col in cat_cols:
    tabla = pd.crosstab(df[col], df[target], normalize="index") * 100
    print(f"\nPorcentaje de riesgo por {col}:")
    display(tabla.round(2))

    riesgo_prom = df.groupby(col)[target].mean().sort_values(ascending=False)

    plt.figure(figsize=(8,4))
    sns.barplot(x=riesgo_prom.index, y=riesgo_prom.values)
    plt.title(f"Proporción de riesgo por {col}")
    plt.ylabel("Proporción de riesgo")
    plt.xticks(rotation=45)
    plt.show()
```

```python
outlier_summary = []

for col in features_num:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = ((df[col] < lower) | (df[col] > upper)).sum()
    outlier_pct = round(outliers / len(df) * 100, 2)

    outlier_summary.append({
        "variable": col,
        "Q1": Q1,
        "Q3": Q3,
        "IQR": IQR,
        "lower_bound": lower,
        "upper_bound": upper,
        "outliers": outliers,
        "outlier_pct": outlier_pct
    })

outlier_df = pd.DataFrame(outlier_summary).sort_values(by="outlier_pct", ascending=False)
display(outlier_df)
```

```python
df_clean = df.copy()
```

```python
# Eliminar identificador
df_clean.drop(columns=["cliente_id"], inplace=True, errors="ignore")

# Eliminar columnas constantes
for col in constant_cols:
    if col in df_clean.columns:
        df_clean.drop(columns=[col], inplace=True)

print("Columnas después de eliminar identificador y constantes:")
print(df_clean.columns.tolist())
```

```python
# One-Hot Encoding para variables categóricas
cat_cols_clean = df_clean.select_dtypes(include=["object"]).columns.tolist()
df_clean = pd.get_dummies(df_clean, columns=cat_cols_clean, drop_first=True, dtype=int)

print("Dataset limpio listo para modelado:")
print(df_clean.shape)
display(df_clean.head())
```
